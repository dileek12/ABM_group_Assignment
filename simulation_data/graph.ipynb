```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("renewable-percentage.csv")
df
```

```python
plt.figure(figsize=(8,5))

# Plot renewable energy
plt.plot(df["[step]"], df["renewable-percentage"], linestyle="-", linewidth=1.5, label="Renewable")

# Plot thermal energy
plt.plot(df["[step]"], 100 - df["renewable-percentage"], linestyle="-", linewidth=1.5, label="Thermal")

plt.xlabel("Simulation Step (h)")
plt.ylabel("Energy Percentage (%)")
plt.title("Renewable vs Thermal Energy Percentage Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
